IDENTIFYING THE OUTLIERS

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
IQR =  q3 - q1
IQR
```

REFINED DATASET BY DROPING THE OUTLIER ROWS

```python
df2 = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR))).any(axis=1)]
df2.shape
```
